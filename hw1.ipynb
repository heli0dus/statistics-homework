```python
import numpy as np
import scipy.stats as stats
import math
```

```python
# дополнительная функция чтобы можно было меньше копипастить для других распределений
gen_sample = lambda n: np.random.default_rng().standard_normal(size=n)
```

```python
eps = 1e-6
answer = []
for i in range(1, 10000, 10):
    sample = gen_sample(i)
    ecdf = stats.ecdf(sample) # эмпирическая функция распределения
    cdf = stats.norm.cdf # честная функция распределения из библиотеки
    x = np.linspace(-100, 100, num=20000)
    f = lambda x: abs(ecdf.cdf.evaluate(x) - cdf(x)) # честно моделируем модуль разности
    answer.append(np.max(f(x))) # супремум в текущей ситуации можно заменить на максимум
```

```python
import matplotlib.pyplot as plt

plt.plot(answer)
```
